```python
    
toronto_data = pd.read_html("toronto_data")
toronto_data
```

# converting html data into dataframe

```python
tdf = toronto_data[0]
tdf.head(15)
```

# removing Borough with not assigned values

```python
tdf = tdf[tdf.Borough != 'Not assigned']
tdf
```

# changing neighbourhood value 'not assigned' to respective Borough value

```python
tdf.replace(to_replace=['Not assigned'], value=["Queen's Park"], inplace= True)
tdf
```

# grouping by postcode

```python
tdf = tdf.groupby(['Postcode']).agg( ','.join)
tdf
```

```python
tdf.reset_index()
```

```python
tdf['Borough'] =tdf['Borough'].astype('str')
tdf['Neighbourhood']= tdf['Neighbourhood'].astype('str')
```

# split the Borough by comma and select the first value to avoid repeatition in Borough column

```python
new = tdf["Borough"].str.split(",", n = 1, expand = True)
tdf['Borough']= new[0]
tdf.head()
```

```python
tdf.reset_index()
```

```python
tdf.shape
```

```python
LL_data = pd.read_csv('LL_data')
```

# adding coordinate values to main data - tdf

```python
Latitude = LL_data['Latitude'].values
Longitude = LL_data['Longitude'].values
tdf['Latitude'] = Latitude
tdf['Longitude'] = Longitude
tdf.head()
```

```python
tdf.reset_index()
```

# Selecting rows that has string Toronto in the Borough column as we are about to map with Toronto city

```python
tdf_toronto = tdf[tdf['Borough'].str.contains('Toronto')]
tdf_toronto
```

```python
tdf_toronto.reset_index()
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="ON_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto City are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(tdf_toronto['Latitude'], tdf_toronto['Longitude'], tdf_toronto['Borough'], tdf_toronto['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
